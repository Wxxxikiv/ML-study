```python
from nltk.probability import ConditionalFreqDist

# 示例句子
sentences = [
    "This is a sample sentence.",
    "ConditionalFreqDist is a useful tool.",
    "Text analysis helps in NLP tasks.",
    "NLTK library is great for text processing.",
]

# 分词并获取句子长度
tokenized_sentences = [sentence.split() for sentence in sentences]
sentence_lengths = [len(tokens) for tokens in tokenized_sentences]
# 创建 ConditionalFreqDist 对象
cfd = ConditionalFreqDist(list(zip(sentence_lengths, sentences)))

# 打印条件下的频率分布
cfd.conditions()
cfd.tabulate(conditions=[5, 6, 7], cumulative=True)
```

```python
import nltk
from nltk.corpus import brown

# 获取新闻和言情文体的句子
news_sentences = brown.sents(categories='news')
romance_sentences = brown.sents(categories='romance')

# 创建函数来分析句子并计算新闻价值和浪漫性
def analyze_day(sentences):
    news_keywords = ['politics', 'economy', 'society']  # 新闻关键词
    romance_keywords = ['love', 'romance', 'affection']  # 浪漫关键词

    news_value = 0
    romance_value = 0

    for sentence in sentences:
        words = [word.lower() for word in sentence]
        news_value += sum(1 for word in words if word in news_keywords)
        romance_value += sum(1 for word in words if word in romance_keywords)

    return news_value, romance_value

# 遍历一周中的每一天，分析新闻和言情文体中的句子
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
best_day = None
best_news_value = 0
best_romance_value = 0

for day in days:
    news, romance = analyze_day(news_sentences if day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else romance_sentences)

    if news > best_news_value and romance > best_romance_value:
        best_day = day
        best_news_value = news
        best_romance_value = romance

print(f"The best day of the week with news value and romance is {best_day}.")
```

## 贝叶斯分类器对名字进行分类

```python
# 特征提取
def gender_features(name: str):
    return {'last_letter': name[-1:], 'last_2_letter': name[-2:], 'last_4_letter': name[-4:], 'first_letter': name[:1], 'first_2_letter': name[:2]}

from nltk.corpus import names
all_names = [(name1, 'male') for name1 in names.words('male.txt')] + [(name2, 'female') for name2 in names.words('female.txt')]
import random
random.shuffle(all_names)
# 划分训练集，开发测试集，测试集
train_set =[(gender_features(name), tag) for name, tag in all_names[600:]]
dev_test_set =[(gender_features(name), tag) for name, tag in all_names[500:600]]
test_set =[(gender_features(name), tag) for name, tag in all_names[:500]]
# 模型训练
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
# 收集错误数据
errors = []
for name, tag in all_names[500:600]:
    guess = classifier.classify(gender_features(name))
    if guess != tag:
        errors.append((tag, guess, name))
if len(errors):
    for (tag, guess, name) in sorted(errors):
        print('correct={:<8} guess={:<8s} name={:<30}'.format(tag, guess, name))
# 模型评估
nltk.classify.accuracy(classifier, test_set)
```

## 对文档进行分类，将影评分为正面或是负面

```python
from nltk.corpus import movie_reviews
movie_reviews.categories()
movie_reviews.fileids('neg')
# 准备数据集
documents = [ (list(movie_reviews.words(fileid)),category)  for category in movie_reviews.categories() for fileid in movie_reviews.fileids(category)]
import random
random.shuffle(documents)
# 提取特征
import nltk
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words)[:2000]
def document_features(document):
    features = {}
    document_words = set(document)
    for word in word_features:
        features['has({})'.format(word)] = (word in document_words)
    return features
# 划分训练集，开发测试集，测试集
train_set =[(document_features(document), category) for document, category in documents[600:]]
dev_test_set =[(document_features(document), category) for document, category in documents[500:600]]
test_set =[(document_features(document), category) for document, category in documents[:500]]
# 模型训练
classifier = nltk.NaiveBayesClassifier.train(train_set)
# 模型评估
nltk.classify.accuracy(classifier, test_set)
```

## 训练一个词性标注器

```python
from nltk.corpus import brown
import nltk
suffix_fdist = nltk.FreqDist()
for word in brown.words():
    suffix_fdist[word[-1:]] += 1
    suffix_fdist[word[-2:]] += 1
    suffix_fdist[word[-3:]] += 1
common_suffix = [suffix for (suffix, count) in suffix_fdist.most_common(100)]

def pos_features(word):
    features = {}
    for suffix in common_suffix:
        features['endswith({})'.format(suffix)] = (word.lower().endswith(suffix))
    return features

tagged_words = brown.tagged_words(categories='news')
featuresets = [(pos_features(word), tag) for (word, tag) in tagged_words]
import random
random.shuffle(featuresets)
test_size = int(len(featuresets)* 0.1)
train_data = featuresets[test_size:]
test_data = featuresets[:test_size]
classifier = nltk.DecisionTreeClassifier.train(train_data)
nltk.classify.accuracy(classifier, test_data)
```

## 基于上下文的特征提取，训练词性标注器
特征提取时传入整个句子和当前词的index，增加当前词的前后词作为新的特征

```python
def pos_features(sentence, i): ❶
    features = {"suffix(1)": sentence[i][-1:],
                "suffix(2)": sentence[i][-2:],
                "suffix(3)": sentence[i][-3:]}
    if i == 0:
        features["prev-word"] = "<START>"
    else:
        features["prev-word"] = sentence[i-1]
    return features
```

## 识别文字含义（RTE）
假设我们有以下文本（T）和假设（H）：
**文本** (Text T): "巴黎是法国的首都，拥有埃菲尔铁塔。"
**假设** (Hypothesis H): "埃菲尔铁塔是一座摩天大楼。"

在这个例子中：
**Word**（词语）： 词语就是文本片段和假设中的具体单词，比如"巴黎"、"法国"、"首都"、"埃菲尔铁塔"等。
**NE**（命名实体）： 在这个例子中，命名实体包括 "巴黎"、"法国" 和 "埃菲尔铁塔"。这是因为它们是具有特定意义的命名的实体，分别表示一个城市、一个国家和一座著名的建筑。

```python
def rte_features(rtepair):
    extractor = nltk.RTEFeatureExtractor(rtepair)
    features = {}
    features['word_overlap'] = len(extractor.overlap('word'))
    features['word_hyp_extra'] = len(extractor.hyp_extra('word'))
    features['ne_overlap'] = len(extractor.overlap('ne'))
    features['ne_hyp_extra'] = len(extractor.hyp_extra('ne'))
    return features

import nltk
rtepair = nltk.corpus.rte.pairs(['rte3_dev.xml'])[33]
extractor = nltk.RTEFeatureExtractor(rtepair)
print(extractor.text_words)
print(extractor.hyp_words)
print(extractor.overlap('word'))
print(extractor.overlap('ne'))
print(extractor.hyp_extra('word'))
# 使用rte提供的分类器训练模型
nltk.classify.rte_classify()
```

## 基于分类器的词块划分器

```python
import nltk
from nltk.corpus import conll2000

def tags_since_dt(sentence, i):
    tags = set()
    for word, pos in sentence[:i]:
        if pos == 'DT':
            tags = set()
        else:
            tags.add(pos)
    return '+'.join(sorted(tags))

def npchunk_features(sentence, i, history):
    word, pos = sentence[i]
    if i == 0:
        prevword, prevpos = "<START>", "<START>"
    else:
        prevword, prevpos = sentence[i-1]
    if i == len(sentence)-1:
        nextword, nextpos = "<END>", "<END>"
    else:
        nextword, nextpos = sentence[i+1]
    return {"pos": pos,
            "word": word,
            "prevpos": prevpos,
            "nextpos": nextpos, 
            "prevpos+pos": "%s+%s" % (prevpos, pos),
            "pos+nextpos": "%s+%s" % (pos, nextpos),
            "tags-since-dt": tags_since_dt(sentence, i)}
    

class ConsecutiveNPChunkTagger(nltk.TaggerI):
    def __init__(self, train_sents):
        train_set = []
        for tagged_sent in train_sents:
            untagged_sent = nltk.tag.untag(tagged_sent)
            history = []
            for i, (word, tag) in enumerate(tagged_sent):
                feature = npchunk_features(untagged_sent, i, history)
                train_set.append((feature, tag))
                history.append(tag)
        nltk.config_megam('C:\Windows\System32\megam.opt')
        self.classifier = nltk.MaxentClassifier.train(train_set, algorithm='megam', trace=1)
    
    def tag(self, sentence):
        history = []
        for i, word, in enumerate(sentence):
            tag = classifier.classify(npchunk_features(sentence, i, history))
            history.append(tag)
        return zip(sentence, history)
        
                
class ConsecutiveNPChunker(nltk.ChunkParserI):
    def __init__(self, train_sents): 
        train_data = [[((w, t), iob) for w,t,iob in nltk.chunk.tree2conlltags(sent)]
                      for sent in train_sents]
        self.tagger = ConsecutiveNPChunkTagger(train_data)

    def parse(self, sentence): 
        pos_tags = [pos for (word,pos) in sentence]
        tagged_pos_tags = self.tagger.tag(pos_tags)
        conlltags = [(w,t,c) for ((w,t),c) in tagged_sents]
        return nltk.chunk.conlltags2tree(conlltags)
    
test_sents = conll2000.chunked_sents('test.txt', chunk_types=['NP'])
train_sents = conll2000.chunked_sents('train.txt', chunk_types=['NP'])
chunker = ConsecutiveNPChunker(train_sents)
print(chunker.accuracy(test_sents))
```
